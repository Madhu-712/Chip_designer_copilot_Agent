#"Product":Design copilot Agent for an semiconductor startup

AI-driven chip design is shaping the semiconductor industry in remarkable ways. The integration of GenAI and Agentic capabilities is just the latest step in the journey to drive greater productivity in an environment of deeper challenges.


Regardless of their level of experience, every designer can take their productivity up across every phase in the chip design cycle. It’s like having an expert guide at your fingertips.This is truly bolstering the productivity of design team.


Copilot would observe what the engineer is working on to provide a more relevant answer. Traditionally, the engineer would research answers on software manuals , seek advice from an experienced co-worker, or even call on an applications engineer—all of which takes much more time and effort.

What’s more, these results could be served up in just minutes.This Copilot is helping the engineer to get the job done faster.

## Creating Agents

```python
from crewai import Crew, Process
from langchain_openai import ChatOpenAI
```

```python
orchestrator_agent = Agent(
    role="Orchestrator",
    goal="Manage the entire AI-assisted chip design workflow for an IT startup, "
         "ensuring seamless integration of AI suggestions and "
         "maintaining design integrity.",
    backstory="As the central control unit, the Orchestrator Agent "
              "orchestrates the interactions between the AI Model "
              "Agent and the Design Agent.  It acts as the bridge "
              "between the user and the AI capabilities, "
              "prioritizing a smooth and efficient design process. "
              "The Orchestrator Agent is responsible for the "
              "overall quality and consistency of the final design.",
    verbose=True,
    allow_delegation=True,
    tools = [scrape_tool, search_tool]

)
```

```python
ai_model_agent = Agent(
    role="AI Model Agent",
    goal="Analyze chip design data using pre-trained AI models "
         "and provide actionable suggestions for bug detection, "
         "optimization, and other improvements.",
    backstory="Trained on massive datasets of chip designs and "
              "verification results, this agent leverages advanced "
              "machine learning to identify potential problems and "
              "suggest improvements that a human designer might "
              "miss.  The AI Model Agent is constantly learning "
              "and adapting, aiming to enhance design efficiency "
              "and quality.",
    verbose=True,
    allow_delegation=True,

)
```

```python
design_agent = Agent(
    role="Design Copilot Agent",
    goal="Act like a Design Assistant or coworker. Evaluate AI-generated suggestions for feasibility, "
         "correctness, and compliance with design constraints. "
         "Ensure the design's integrity and functionality.",
    backstory="This agent possesses deep expertise in EDA (Electronic design automation)tools and "
              "the intricacies of chip design,chip architecture.  It acts as a "
              "gatekeeper, carefully reviewing AI suggestions "
              "before they are integrated.  The Design Agent "
              "prioritizes design quality and reliability, "
              "preventing the introduction of errors or "
              "undesirable trade-offs.",
   verbose=True,
   allow_delegation=True,
)
```

## Creating Tasks

```python
orchestrator_task = Task(
    description=(
        "Manage the entire AI-assisted chip design upgrade workflow an IT startup. "
        "Receive user input (design data and instructions), "
        "coordinate the AI Model Agent's analysis, "
        "send suggestions to the Design Agent for validation, "
        "integrate approved changes, and display results to the user. "
        "Ensure seamless collaboration between all agents and maintain "
        "design integrity."
    ),
    expected_output=(
        "A successfully upgraded chip design incorporating AI-suggested "
        "improvements, along with a report detailing the changes made(from traditional  to AI enabled chip)"
        "and any feedback from the Design Agent."
    ),
    agent=orchestrator_agent,
)
```

```python
ai_model_task = Task(
    description=(
        "Analyze the provided chip design data using pre-trained AI "
        "models to identify potential bugs, suggest optimizations "
        "Power consumption, area, performance,reliability and cost), and provide predictions "
        "Eg.,1.Performance-speed,latency,throughput"
        "2.Power consumption-efficiency,dynamic and static power)."
        "3.Manufacturing complexity  and yeild."
        "4.Reliability-Robustness and failure rate"
        "5.Other factors-Area,pincount,security"
        "Prioritize actionable insights "
        "relevant to the user's specified upgrade goals."
    ),
    expected_output=(
        "A structured report including flagged potential bugs, specific "
        "optimization suggestions (with justifications), and predictions "
        "regarding the impact of the suggestions. The output should be "
        "formatted for easy integration with EDA tools."
    ),
    agent=ai_model_agent,
)
```

```python
design_agent_task = Task(
    description=(
        "Act like a design copilot/coworker/design assistant."
        "Review AI-generated suggestions for the chip design upgrade. "
        "Validate the feasibility, correctness, and impact of each "
        "suggestion within the context of the existing design. Ensure "
        "that all proposed changes comply with design constraints "
        "(Performance,Powerconsumption,Manufacturing complexity,Reliability and other factors like power, area). Provide feedback on the AI suggestions "
        "and highlight potential risks or conflicts."
    ),
    expected_output=(
        "A report detailing the validation process for each AI suggestion, "
        "including acceptance/rejection decisions and justifications.  If "
        "suggestions are modified, provide the updated version.  Include "
        "any warnings or risks associated with the proposed changes."
    ),
    agent=design_agent,
)
```

## Creating the Crew
- The `Process` class helps to delegate the workflow to the Agents (kind of like a Manager at work)
- In the example below, it will run this hierarchically.
- `manager_llm` lets you choose the "manager" LLM you want to use.

```python
from crewai import Crew, Process
Crew=Crew(
   agents=[orchestrator_agent,
           ai_model_agent,
           design_agent],
   tasks=[orchestrator_task,
          ai_model_task,
          design_agent_task],
   manager_llm=ChatOpenAI(model="gpt-3.5-turbo",
                           temperature=0.5),

   process=Process.hierarchical,
   verbose=True
)
```

#Create a verilog file having HDL(Hardware description language)

```python
%%writefile verilog.v
module register_file (
    input clk,
    input rst,
    input write_en,
    input [2:0] address_read,
    input [2:0] address_write,
    input [7:0] data_in,
    output reg [7:0] data_out_r
);
reg [7:0] reg_file [6:0];
always @(posedge clk or posedge rst) begin
    if (rst) begin
        reg_file[0] <= 8'b0;
        reg_file[1] <= 8'b0;
        reg_file[2] <= 8'b0;
        reg_file[3] <= 8'b0;
        reg_file[4] <= 8'b0;
        reg_file[5] <= 8'b0;
        reg_file[6] <= 8'b0;
    end else begin
        if (write_en) begin
            reg_file[address_write] <= data_in;
        end
    end
end
always @(*) begin
    data_out_r = reg_file[address_read];
end
endmodule
```

#Streamlit code

```python

#streamlit code

import streamlit as st
from crewai import Crew, Agent, Task, Process
from langchain_openai import ChatOpenAI
from crewai_tools import ScrapeWebsiteTool, SerperDevTool

# Initialize LLM
manager_llm = ChatOpenAI(model="gpt-3.5-turbo", temperature=0.5)

# Initialize tools
search_tool = SerperDevTool()
scrape_tool = ScrapeWebsiteTool()

# Initialize agents
orchestrator_agent = Agent(
    role="Orchestrator",
    goal="Manage the entire AI-assisted chip design workflow for an IT startup, "
         "ensuring seamless integration of AI suggestions and "
         "maintaining design integrity.",
    backstory="As the central control unit, the Orchestrator Agent "
         "orchestrates the interactions between the AI Model "
         "Agent and the Design Agent.  It acts as the bridge "
         "between the user and the AI capabilities, "
         "prioritizing a smooth and efficient design process. "
         "The Orchestrator Agent is responsible for the "
         "overall quality and consistency of the final design.",
    verbose=True,
    allow_delegation=True,
    tools=[scrape_tool, search_tool],
)

ai_model_agent = Agent(
    role="AI Model Agent",
    goal="Analyze chip design data using pre-trained AI models "
         "and provide actionable suggestions for bug detection, "
         "optimization, and other improvements.",
    backstory="Trained on massive datasets of chip designs and "
              "verification results, this agent leverages advanced "
              "machine learning to identify potential problems and "
              "suggest improvements that a human designer might "
              "miss.  The AI Model Agent is constantly learning "
              "and adapting, aiming to enhance design efficiency "
              "and quality.",
    verbose=True,
    allow_delegation=True,

)

design_agent = Agent(
     role="Design Copilot Agent",    goal="Act like a Design Assistant or coworker. Evaluate AI-generated suggestions for feasibility, "
         "correctness, and compliance with design constraints. "
         "Ensure the design's integrity and functionality.",
     backstory="This agent possesses deep expertise in EDA (Electronic design automation)tools and "
              "the intricacies of chip design,chip architecture.  It acts as a "
              "gatekeeper, carefully reviewing AI suggestions "
              "before they are integrated.  The Design Agent "
              "prioritizes design quality and reliability, "
              "preventing the introduction of errors or "
              "undesirable trade-offs.",
    verbose=True,
    allow_delegation=True,
)

# Initialize tasks
orchestrator_task = Task(
    description=(
        "Manage the entire AI-assisted chip design upgrade workflow an IT startup. "
        "Receive user input (design data and instructions), "
        "coordinate the AI Model Agent's analysis, "
        "send suggestions to the Design Agent for validation, "
        "integrate approved changes, and display results to the user. "
        "Ensure seamless collaboration between all agents and maintain "
        "design integrity."
    ),
    expected_output=(
        "A successfully upgraded chip design incorporating AI-suggested "
        "improvements, along with a report detailing the changes made(from traditional  to AI enabled chip)"
        "and any feedback from the Design Agent."
    ),
    agent=orchestrator_agent,
)

ai_model_task = Task(
    description=(
        "Analyze the provided chip design data using pre-trained AI "
        "models to identify potential bugs, suggest optimizations "
        "Power consumption, area, performance,reliability and cost), and provide predictions "
        "Eg.,1.Performance-speed,latency,throughput"
        "2.Power consumption-efficiency,dynamic and static power)."
        "3.Manufacturing complexity  and yeild."
        "4.Reliability-Robustness and failure rate"
        "5.Other factors-Area,pincount,security"
        "Prioritize actionable insights "
        "relevant to the user's specified upgrade goals."
    ),
    expected_output=(
        "A structured report including flagged potential bugs, specific "
        "optimization suggestions (with justifications), and predictions "
        "regarding the impact of the suggestions. The output should be "
        "formatted for easy integration with EDA tools."
    ),
    agent=ai_model_agent,
)


design_agent_task = Task(
    description=(
        "Act like a design copilot/coworker/design assistant."
        "Review AI-generated suggestions for the chip design upgrade. "
        "Validate the feasibility, correctness, and impact of each "
        "suggestion within the context of the existing design. Ensure "
        "that all proposed changes comply with design constraints "
        "(Performance,Powerconsumption,Manufacturing complexity,Reliability and other factors like power, area). Provide feedback on the AI suggestions "
        "and highlight potential risks or conflicts."
    ),
    expected_output=(
        "A report detailing the validation process for each AI suggestion, "
        "including acceptance/rejection decisions and justifications.  If "
        "suggestions are modified, provide the updated version.  Include "
        "any warnings or risks associated with the proposed changes."
    ),
    agent=design_agent,
)


# Initialize Crew
Crew = Crew(
    agents=[orchestrator_agent, ai_model_agent, design_agent],
    tasks=[orchestrator_task, ai_model_task, design_agent_task],
    manager_llm=manager_llm,
    process=Process.hierarchical,
    verbose=True,
)

# Streamlit App
st.title("AI-Assisted Chip Design")

# User Inputs (file uploader, multiselect, number inputs, text inputs)
design_data = st.file_uploader("Upload Verilog Design File", type=["v"])
upgrade_goals = st.multiselect(
    "Select Upgrade Goals",
    ["reduce_power_consumption", "improve_performance", "Reliability", "cost"],
)
constraints = {
    "power_budget": st.number_input("Power Budget (Watts)", value=1.5),
    "area_limit": st.number_input("Area Limit (mm^2)", value=100),
    "performance_target": st.number_input("Performance Target (GHz)", value=10),
}

# Optional AI Model Preferences
ai_model_preferences = {
    "bug_detection_model": st.text_input("Bug Detection Model", value="model_A"),
    "optimization_model": st.text_input("Optimization Model", value="model_B"),
}
# Run CrewAI
if st.button("Run AI Analysis"):
    # Create inputs dictionary
    inputs = {        "design_data": design_data.read().decode("utf-8") if design_data else None,  # Handle file upload
        "upgrade_goals": upgrade_goals,
        "constraints": constraints,
        "ai_model_preferences": ai_model_preferences,
        "eda_tool": eda_tool,
    }
    with st.spinner("Running AI analysis..."):
         result = Crew.kickoff(inputs=inputs)
         st.markdown(result)
```
